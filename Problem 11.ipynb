```python
def vertical():
    p = 0
    for i in range(0,340): #only running through the first 17 rows. Need to be mindful that we will be out of index if we run through all rows.
        n = int(list[i])*int(list[i+20])*int(list[i+40])*int(list[i+60])
        if n > p:
            p = n
    return(p)
vertical()
```

```python
def horizontal():
    s = 0
    for i in range (0, 20): #ith row
        for j in range (0,17): #jth column
            t = i*20 + j  
            k = int(list[t])*int(list[t+1])*int(list[t+2])*int(list[t+3])
            if k > s:
                s = k
    return(s)
horizontal()
```

```python
def south_eastward_diagonal():
    r = 0
    for i in range (0, 17): #ith row
        for j in range (0,17): #jth column
            w = i*20 + j
            q = int(list[w])*int(list[w+21])*int(list[w+42])*int(list[w+63])
            if q > r:
                r = q
    return(r)
south_eastward_diagonal()
```

```python
def south_westward_diagonal():
    m = 0
    for i in range (0, 17): #ith row
        for j in range (3,20): #jth column
            u = i*20 + j
            v = int(list[u])*int(list[u+19])*int(list[u+38])*int(list[u+57])
            if v > m:
                m = v
    return(m)
south_westward_diagonal()
```
