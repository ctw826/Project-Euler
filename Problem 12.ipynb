```python
import math

def triangle_number():
    i = 1
    while True:
        yield i*(i+1)/2
        i+=1

        #Sum of natural numbers = n(n+1)/2
```

```python
#brute force
def first_triangle_number_over_five_hundred_divisors(triangle):
    for n in triangle:
        sqrt = math.sqrt(n)
        int_sqrt = int(sqrt)
        number_of_divisor = 2
        for t in range (2, int_sqrt+1):
            if n%t==0:
                if t == sqrt:
                    number_of_divisor += 1
                else:
                    number_of_divisor += 2
        if number_of_divisor >= 500:
            return n
        
        
```

```python
first_triangle_number_over_five_hundred_divisors(triangle_number())
```
